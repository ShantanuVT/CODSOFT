```python
import math
board = [' ' for _ in range(9)]
human = 'X'
ai = 'O'
print("=" * 40)
print("  Welcome to Tic-Tac-Toe!")
print("=" * 40)
print("\nYou are X, AI is O")
first = input("\nWho goes first? (1: You, 2: AI): ").strip()
current_player = human if first == '1' else ai
while True:
    print("\n")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---|---|---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---|---|---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print("\n")
    print("Positions:")
    print(" 0 | 1 | 2 ")
    print("---|---|---")
    print(" 3 | 4 | 5 ")
    print("---|---|---")
    print(" 6 | 7 | 8 ")
    print("\n")
    if current_player == human:
        try:
            move = int(input("Enter your move (0-8): "))
            if move not in range(9):
                print("Invalid position! Choose 0-8.")
                continue
            if board[move] != ' ':
                print("That position is taken! Try again.")
                continue
            board[move] = human
        except (ValueError, IndexError):
            print("Invalid input! Enter a number 0-8.")
            continue
    else:
        print("AI is thinking...")
        best_move = -1
        best_value = -math.inf
        available_moves = [i for i, spot in enumerate(board) if spot == ' ']
        for move in available_moves:
            board[move] = ai
            stack = [(0, False, -math.inf, math.inf, list(board), None)]
            move_value = None
            while stack:
                depth, is_maximizing, alpha, beta, current_board, result = stack.pop()
                if result is not None:
                    if is_maximizing:
                        alpha = max(alpha, result)
                    else:
                        beta = min(beta, result)
                    continue
                win_combinations = [
                    [0, 1, 2], [3, 4, 5], [6, 7, 8],
                    [0, 3, 6], [1, 4, 7], [2, 5, 8],
                    [0, 4, 8], [2, 4, 6]
                ]
                ai_wins = any(all(current_board[i] == ai for i in combo) for combo in win_combinations)
                human_wins = any(all(current_board[i] == human for i in combo) for combo in win_combinations)
                board_full = ' ' not in current_board
                if ai_wins:
                    move_value = 10 - depth
                    continue
                if human_wins:
                    move_value = depth - 10
                    continue
                if board_full:
                    move_value = 0
                    continue
                available = [i for i, spot in enumerate(current_board) if spot == ' ']
                if is_maximizing:
                    max_eval = -math.inf
                    for pos in available:
                        new_board = current_board[:]
                        new_board[pos] = ai
                        ai_wins_new = any(all(new_board[i] == ai for i in combo) for combo in win_combinations)
                        human_wins_new = any(all(new_board[i] == human for i in combo) for combo in win_combinations)
                        board_full_new = ' ' not in new_board
                        
                        if ai_wins_new:
                            eval_score = 10 - (depth + 1)
                        elif human_wins_new:
                            eval_score = (depth + 1) - 10
                        elif board_full_new:
                            eval_score = 0
                        else:
                            available_new = [i for i, spot in enumerate(new_board) if spot == ' ']
                            min_eval = math.inf
                            for pos2 in available_new:
                                new_board2 = new_board[:]
                                new_board2[pos2] = human
                                ai_wins_new2 = any(all(new_board2[i] == ai for i in combo) for combo in win_combinations)
                                human_wins_new2 = any(all(new_board2[i] == human for i in combo) for combo in win_combinations)
                                board_full_new2 = ' ' not in new_board2
                                if ai_wins_new2:
                                    eval_score2 = 10 - (depth + 2)
                                elif human_wins_new2:
                                    eval_score2 = (depth + 2) - 10
                                elif board_full_new2:
                                    eval_score2 = 0
                                else:
                                    eval_score2 = 0
                                min_eval = min(min_eval, eval_score2)
                                if min_eval <= alpha:
                                    break
                            eval_score = min_eval
                        max_eval = max(max_eval, eval_score)
                        alpha = max(alpha, eval_score)
                        if beta <= alpha:
                            break
                    move_value = max_eval
                else:
                    min_eval = math.inf
                    for pos in available:
                        new_board = current_board[:]
                        new_board[pos] = human
                        ai_wins_new = any(all(new_board[i] == ai for i in combo) for combo in win_combinations)
                        human_wins_new = any(all(new_board[i] == human for i in combo) for combo in win_combinations)
                        board_full_new = ' ' not in new_board
                        if ai_wins_new:
                            eval_score = 10 - (depth + 1)
                        elif human_wins_new:
                            eval_score = (depth + 1) - 10
                        elif board_full_new:
                            eval_score = 0
                        else:
                            eval_score = 0
                        min_eval = min(min_eval, eval_score)
                        beta = min(beta, eval_score)
                        if beta <= alpha:
                            break
                    move_value = min_eval
            board[move] = ' '
            if move_value > best_value:
                best_value = move_value
                best_move = move
        board[best_move] = ai
        print(f"AI chose position {best_move}")
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        [0, 4, 8], [2, 4, 6]
    ]
    winner_found = False
    for combo in win_combinations:
        if all(board[i] == current_player for i in combo):
            print("\n")
            print(f" {board[0]} | {board[1]} | {board[2]} ")
            print("---|---|---")
            print(f" {board[3]} | {board[4]} | {board[5]} ")
            print("---|---|---")
            print(f" {board[6]} | {board[7]} | {board[8]} ")
            print("\n")
            if current_player == human:
                print("Congratulations! You win!")
            else:
                print("AI wins! Better luck next time.")
            winner_found = True
            break
    if winner_found:
        break
    if ' ' not in board:
        print("\n")
        print(f" {board[0]} | {board[1]} | {board[2]} ")
        print("---|---|---")
        print(f" {board[3]} | {board[4]} | {board[5]} ")
        print("---|---|---")
        print(f" {board[6]} | {board[7]} | {board[8]} ")
        print("\n")
        print("It's a draw!")
        break
    current_player = ai if current_player == human else human
play_again = input("\nPlay again? (y/n): ").lower()
if play_again == 'y':
    import os
    
    os.system('python' + __file__)
else:
    print("\nThanks for playing! Goodbye!")
```
